```python
import os 
import pandas as pd
import numpy as np

os.chdir("/home/t326h379/OGP")
Header_name = ["Amino_Acid"]

col_of_feature = [i for i in range(1,1025)]

Header_name = Header_name + col_of_feature

def open_file(label_i, protein_id, csv_file_name):  

    
    df = pd.read_csv(csv_file_name, header=None)
    
    df.columns = Header_name

    
        

    with open(csv_file_name) as outfile:
        feature_vector = []
        AA = []
        lines = outfile.readlines()
        for line in lines:
            list = line.split(",")
            AA.append(list[0])
           
        AA = ''.join(AA)
       
        
        index = label_i-1
        
        if (AA[index] == "S" or AA[index] == "T"):
        

            matrix = df.iloc[index]
            grabbed_matrix =  np.array(matrix)


            flatten_value = grabbed_matrix.ravel()

            feature_vector = flatten_value.tolist()


            listToStr = ','.join([str(elem) for elem in feature_vector])
            



            with open('Feature_Extraction_July_30_Sprint_Gly_Negative_Independent_Test_3376_.txt','a+') as outfile:               
                outfile.write(str(label_i))
                outfile.write(",")
                outfile.write(str(protein_id))
                outfile.write(",")
                outfile.write(listToStr)
                outfile.write("\n")
                
with open("Spring_Gly_O_linked_Testing_Negative.txt") as subash:
    for line in subash:
        x = line.split()
        PID = x[0]
        Position = int(x[1])
        csv_file_name = PID+"_Prot_Trans_.csv"
        label_i = Position
        try:
            open_file(Position, PID, csv_file_name)
        except:
            print("File Not Found!!!")
            pass
```
